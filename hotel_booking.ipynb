```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('hotel_bookings.csv')
print("Shape:", df.shape)
display(df.head(10))
df.info()

missing = df.isna().sum().sort_values(ascending=False)
print(missing.head(40))

missing_nonzero = missing[missing>0]
if len(missing_nonzero)>0:
    plt.figure(figsize=(10,4))
    missing_nonzero.head(40).plot(kind='bar')
    plt.title('Missing values per column (top 40)')
    plt.tight_layout()
    plt.show()
# describe
num = df.select_dtypes(include=['number'])
display(num.describe().transpose())
if 'adr' in df.columns:
    plt.figure(); plt.boxplot(df['adr'].dropna()); plt.title('Boxplot of adr'); plt.show()
if 'lead_time' in df.columns:
    plt.figure(); plt.boxplot(df['lead_time'].dropna()); plt.title('Boxplot of lead_time'); plt.show()

cats = ['hotel','meal','market_segment','distribution_channel','reserved_room_type','deposit_type','customer_type']
for c in cats:
    if c in df.columns:
        print(f"\n-- {c} --")
        print(df[c].value_counts(dropna=False).head(10))
if 'country' in df.columns:
    print("\nTop countries:")
    print(df['country'].value_counts().head(20))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
df = pd.read_csv('hotel_bookings.csv')
df = df.drop_duplicates()

df['agent'] = pd.to_numeric(df['agent'], errors='coerce').fillna(0).astype(int)
df['company'] = pd.to_numeric(df['company'], errors='coerce').fillna(0).astype(int)

df['country'] = df['country'].fillna(df['country'].mode().iloc[0])

df['children'] = pd.to_numeric(df['children'], errors='coerce')
df['children'] = df['children'].fillna(df['children'].median()).astype(int)

df['babies'] = pd.to_numeric(df['babies'], errors='coerce').fillna(0).astype(int)
df['adults'] = pd.to_numeric(df['adults'], errors='coerce').fillna(1).astype(int)

df['adr'] = pd.to_numeric(df['adr'], errors='coerce')
df.loc[df['adr'] < 0, 'adr'] = np.nan
df['adr'] = df['adr'].clip(upper=1000)

df['total_guests'] = df['adults'] + df['children'] + df['babies']
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['is_family'] = ((df['children'] > 0) | (df['babies'] > 0)).astype(int)
for col in ['reservation_status','reservation_status_date']:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

df.to_csv('hotel_bookings_cleaned.csv', index=False)

train, test = train_test_split(df, test_size=0.2, random_state=42,
                               stratify=df['is_canceled'])
train.to_csv('hotel_bookings_train.csv', index=False)
test.to_csv('hotel_bookings_test.csv', index=False)
print("Cleaned data saved as hotel_bookings_cleaned.csv")
print("Train/Test splits saved.")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('hotel_bookings_cleaned.csv')

cat_cols = df.select_dtypes(include=['object']).columns.tolist()
low_card_threshold = 10
low_card = [c for c in cat_cols if df[c].nunique() <= low_card_threshold]
high_card = [c for c in cat_cols if df[c].nunique() > low_card_threshold]
for c in high_card:
    freq = df[c].value_counts(normalize=True)
    rare = freq[freq < 0.01].index
    if len(rare) > 0:
        df[c] = df[c].replace(rare, 'Other')
    vc = df[c].value_counts(normalize=True)
    df[c + '_freq_enc'] = df[c].map(vc.to_dict())
if low_card:
    df = pd.get_dummies(df, columns=low_card, drop_first=False)

for c in high_card:
    if c in df.columns:
        df = df.drop(columns=[c])

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
exclude = ['is_canceled'] if 'is_canceled' in df.columns else []
continuous_cols = [c for c in num_cols if df[c].nunique() > 5 and c not in exclude]
if continuous_cols:
    scaler = StandardScaler()
    df[continuous_cols] = scaler.fit_transform(df[continuous_cols])

df.to_csv('hotel_bookings_preprocessed.csv', index=False)
print("Saved preprocessed as hotel_bookings_preprocessed.csv (shape: {})".format(df.shape))
```
